```python
sql = f'''WITH dates(month) AS (
  VALUES(({min}))
  UNION ALL
  SELECT date(month, '+1 month')
  FROM dates
  WHERE month < ({max})
),
trans_month as (
select date(t.dt,'start of month') as month,
count(*) as transactions_cnt,
sum(t.amount) as amount_sum
from client_transaction t
group by 1
order by 1
)

SELECT t.month,
coalesce(tm.transactions_cnt,0) as transactions_cnt,
coalesce(tm.amount_sum,0) as amount_sum

FROM dates t
left join trans_month tm on tm.month=t.month
order by t.month
'''
```

```python
select(sql)
```

```python
sql = f'''WITH dates(month) AS (
  VALUES(({min}))
  UNION ALL
  SELECT date(month, '+1 month')
  FROM dates
  WHERE month < ({max})
),
clients as (
  select distinct t.client_id from german_credit t
),

clients_month as (SELECT t.month,c.client_id FROM dates t
join clients c on 1=1),

trans_month as (
select date(t.dt,'start of month') as month,
t.client_id,
count(*) as transactions_cnt,
sum(t.amount) as amount_sum
from client_transaction t
group by 1,2
order by 1
),
client_trans_month as (
SELECT t.*,
tm.transactions_cnt,
tm.amount_sum,
1 as user
,case when tm.transactions_cnt then 1 else 0 end active

from clients_month t left join trans_month tm on
t.client_id=tm.client_id and t.month=tm.month
)

select t.month, sum(t.user) as user_cnt, sum(t.amount_sum) as amount_sum,
sum(t.active) as active_cnt
from client_trans_month t
group by t.month
'''
```

```python
t = select(sql)
```
